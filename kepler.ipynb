```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
```

```python
fig, ax = plt.subplots()
ax.scatter(berlinplan['activity'], berlinplan['habitants'], s=berlinplan['surface']*0.00004,alpha=0.5)
#c=close, s=volume, 
#ax.set_xlabel(r'$\Delta_i$', fontsize=15)
#ax.set_ylabel(r'$\Delta_{i+1}$', fontsize=15)
ax.set_title('Berlin Planning Zones')

ax.grid(True)
fig.tight_layout()

plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(18, 10))
flatui = ['#9e0142' ,'#d53e4f',"#95a5a6", "#f46d43", "#fdae61", "#fee08b","#1a1a1a", "#e6f598", "#abdda4", "#66c2a5", "#3288bd",
          '#5e4fa2']
ax = sns.scatterplot(x="activity", y="habitants",
                     hue='BEZIRK', size=berlinplan['surface'].values,
                     sizes=(20, 3000), hue_norm=(0, 50),alpha=0.9,palette=flatui, data=berlinplan)
ax.set_title("Berlin Urban Economical Activities x Population")
plt.xlabel("Urban economical Activities")
plt.ylabel("Population")
h,l = ax.get_legend_handles_labels()
ax.legend(h[:12], l[:12],loc='upper right',ncol=2,shadow=True, borderpad=1,title="Districts",scatterpoints=1,prop={'size': 12},
          bbox_to_anchor=(1.15, 1.01))

plt.savefig('D:/urbban/berlin_scatter.png')
```

```python
sns.set_context("paper", font_scale=3, rc={"font.size":12,"axes.labelsize":25})

plt.figure(figsize=(35, 10))
plt.xticks(rotation=45)
plt.subplots_adjust(bottom=0.15)
ax = sns.swarmplot(x='BEZIRK', y="activity", palette=flatui,size=12,data=berlinplan)
plt.xlabel("District")
plt.ylabel("Urban economical Activities")
plt.title("Economical Activities Swarnplot in Planning Zones")
```
