## Create Altair scatterplot with barchart about Berlin urban activities


```python
import pandas as pd
import altair as alt
import geopandas as gpd
```

```python
berlinplan = gpd.read_file('D:/urbban/Planungsraum_EPSG_25833_well.shp',encoding='utf-8')  
berlinplan = gpd.GeoDataFrame(berlinplan,geometry='geometry',crs={'init':'epsg:25833'})
berlinplan['surface']= berlinplan['surface'].round(1)

```

```python
berlinplan.groupby('BEZIRK')['activity'].sum()
```

```python
The plot

scale = alt.Scale(domain=['Mitte', 'Friedrichshain-Kreuzberg', 'Pankow',
       'Charlottenburg-Wilmersdorf', 'Spandau', 'Steglitz-Zehlendorf',
       'Tempelhof-Schöneberg', 'Neukölln', 'Treptow-Köpenick',
       'Marzahn-Hellersdorf', 'Lichtenberg', 'Reinickendorf'],
        range=['#9e0142' ,'#d53e4f',"#95a5a6", "#f46d43", "#fdae61", "#fee08b","#1a1a1a", "#e6f598", "#abdda4", "#66c2a5", "#3288bd",
          '#5e4fa2'])
color = alt.Color('BEZIRK:N', scale=scale)

# two selections:
# - a brush that is active on the top panel
# - a multi-click that is active on the bottom panel
brush = alt.selection_interval(encodings=['x'])
click = alt.selection_multi(encodings=['color'])

# Top panel is scatter plot 
points = alt.Chart().mark_point(filled=True).encode(
    alt.X('activity:Q', title='Activities'),
    alt.Y('habitants:Q',
        title='Habitants'
        )
    ,
    color=alt.condition(brush, color, alt.value('lightgray')),
        tooltip = [alt.Tooltip('BEZIRK:N'),
               alt.Tooltip('activity:Q'),
               alt.Tooltip('habitants:Q'),
               alt.Tooltip('surface:Q')],
    size=alt.Size('surface:Q', scale=alt.Scale(type= "quantile",range=[100,200,300,500,700,1000])),
    opacity = alt.OpacityValue(0.5)
).properties(
    width=700,
    height=500
).add_selection(
    brush
).transform_filter(
    click
)
# Bottom panel is a bar chart 
bars = alt.Chart().mark_bar(filled=True).encode(
  alt.X( 'sum(activity):Q',title='Sum of Activities'),
  alt.Y( 'BEZIRK:N',title='Districts'),
    color=alt.condition(click, color, alt.value('lightgray')),
    tooltip = [alt.Tooltip('BEZIRK:N'),
               alt.Tooltip('sum(activity):Q')],
    
    opacity = alt.OpacityValue(0.5),
).transform_filter(
    brush
).properties(
    width=550,
).add_selection(
    click
)

ost = alt.vconcat(
    
    points,
    bars,
    data=berlinplan,
    title="Berlin Districts",


)
```
